```python
import pandas as pd
import os
import re
```

```python
df = pd.read_csv('all_tpm.csv')
df = df.set_index('Gene')
df.head()
```

```python
# Case 1:
write_path = os.path.join(os.path.abspath('.'),'1_WTIL_WTnoIL','input')

header_map = dict(
WT_t1='WT_IL_Early',
WT_t2='WT_IL_Mid',
MT_t1='WT_NoIL_Early',
MT_t2='WT_NoIL_Mid')

def write_fun(header_map, write_path):
    col_ids = [s for s in df.columns for l in list(header_map.values()) if l in s]
    subdf = df[col_ids]
    name_dict = {v: k for k, v in header_map.items()}
    subdf = subdf.rename(columns=lambda x: x.replace(x[:-5],name_dict[x[:-5]]))
    subdf.to_csv(os.path.join(write_path,'tpm.csv'))

    hm = pd.DataFrame({'INPUT_ID':list(header_map.keys()), 'ORIGINAL_ID':list(header_map.values())})
    hm.to_csv(os.path.join(write_path,'header_map.csv'), index=False)
    
write_fun(header_map, write_path)
```

```python
# Case 2:
write_path = os.path.join(os.path.abspath('.'),'2_MTIL_MTnoIL','input')

header_map = dict(
WT_t1='MT_IL_Early',
WT_t2='MT_IL_Mid',
MT_t1='MT_NoIL_Early',
MT_t2='MT_NoIL_Mid')

write_fun(header_map, write_path)
```

```python
# Case 3:
write_path = os.path.join(os.path.abspath('.'),'3_MTnoIL_WTnoIL','input')

header_map = dict(
WT_t1='MT_NoIL_Early',
WT_t2='MT_NoIL_Mid',
MT_t1='WT_NoIL_Early',
MT_t2='WT_NoIL_Mid')

write_fun(header_map, write_path)
```
